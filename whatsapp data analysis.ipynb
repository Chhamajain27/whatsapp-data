### import the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

## Import The data

```python
f=open('WhatsApp Chat .txt',encoding='utf-8')   ##File pointer  ## it open the file in r mode
```

```python
data=f.read()
data
```

```python
## Checking the type of data
type(data)
```

```python
print(data)
```

```python
l=data.split('_')
l
```

```python
pattern="\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
messages=re.split(pattern,data)
messages
```

```python
type(messages)
```

```python
messages=messages[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
df
```

```python
df.info()
```

```python
df["message_date"].unique()
```

```python
## since date is object
df['message_date']=pd.to_datetime(df['message_date'],format="%d/%m/%Y, %H:%M - ")
```

```python
df
```

```python
df.shape
```

```python
df.head()
```

## splitting the message date column and extracting the year, month,day,hour,day_name etc.

```python
df["year"]=df["message_date"].dt.year
df["month"]=df["message_date"].dt.month_name()
df["day"]=df["message_date"].dt.day
df["hour"]=df["message_date"].dt.hour
df["hour"]=df["message_date"].dt.month
df["day_name"]=df["message_date"].dt.day_name()
df["Date"]=df["message_date"].dt.date
df
```

### Drop the message date Column

```python
df.drop(columns=['message_date'])
```

```python
users=[]
messages=[]
for i in df['user_message']:
    print(i)
    if"changed the subject" not in i:
        li=i.split(":",maxsplit=1)
    else:
        li=[]
        li.append(i)
    #print(li)
    if len(li)>=2:
        users.append(li[0])
        messages.append(li[1])
    else:
        users.append("group notification")
        messages.append(li[0])
df['user']=users
df['message']=messages
df
```

```python
df.drop(columns=['user_message'],inplace=True)
```

```python
from urlextract import URLExtract
```

```python
extractor= URLExtract()
word=[]
for i in words:
    if len(extractor.find_urls(i))==0:
        word.append(i)
word
```

```python
s=[]
for i in word:
    if '<Media'!=i and 'omitted>'!=i :
        s.append(i)
s
    
```
